```python
from collections import Counter

def parse_molecule(molecule):

    array = [[]]

    for token in map(str, molecule):
        if token.isalpha() and token.istitle():
            last = [token]
            upper = token
            array[-1].append(token)
        elif token.isalpha():
            last = upper + token
            array[-1] = [last]
        elif token.isdigit():
            array[-1].extend(last*(int(token)-1))
        elif token == '(' or token == '[':
            array.append([])
        elif token == ')' or token == ']':
            last = array.pop()
            array[-1].extend(last)

    return dict(Counter(array[-1]))
```

```python
water = 'H2O'
parse_molecule(water)
```

```python
p2=pd.DataFrame(pd.read_csv(r'perovskite-ion.csv'))
p2=pd.DataFrame(p2['Name'])
mats = p2['Name'].values.tolist()
mats
```

```python
import re

def tokenize_formula(formula):
    return re.findall('[A-Z][a-z]?|\d+|.', formula)


df1=pd.DataFrame()

# for formula in mats:
#     print("\t", tokenize_formula(formula))
#     # df2=df1.append(pd.DataFrame({'code':[formula]}),ignore_index=True)
import warnings
warnings.filterwarnings("ignore")

for formula in mats:
    tokenized = tokenize_formula(formula)
    df1 = df1.append({'Formula': formula, 'Tokenized_Formula': tokenized}, ignore_index=True)
    
df1
```

```python
# 使用 pd.Series() 将列表拆分成不同的列
df_elements = df1['Tokenized_Formula'].apply(pd.Series)

# 重命名列名
df_elements.columns = [f'Element_{i+1}' for i in range(df_elements.shape[1])]

# 合并到原始 DataFrame
df2 = pd.concat([df1, df_elements], axis=1)
df2
```
